```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#reading csv file data using padas
df=pd.read_csv("homeprices.csv")
#printing the read data
df
```

```python
%matplotlib inline
plt.xlabel("area(sq ft)")
plt.ylabel("price (US$)")

#plotting the data
plt.scatter(df.area,df.price,color="red")
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
#plotting the data of csv file
%matplotlib inline
plt.xlabel("area(sq ft)")
plt.ylabel("price (US$)")
plt.scatter(df.area,df.price,color="red")

#plotting the stright that best fits
plt.plot(df.area,reg.predict(df[['area']]),color="blue")
```

```python
#predictingg the price of the house with 3300 sq feet
reg.predict([[3300]])
```

```python
#reg coefficient
reg.coef_
```

```python
#reg intercept
reg.intercept_
```

```python
#predicting the home prices which are in the csv file

area_df = pd.read_csv("areas.csv")
area_df.head(3)
```

```python
predict_price= reg.predict(area_df)
predict_price 
```

```python
area_df['prices']=predict_price
area_df
```

```python
area_df.to_csv("prediction.csv")
```
